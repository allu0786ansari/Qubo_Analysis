# Qubolite

```python
import numpy as np
import os
import itertools
from qubolite import qubo
from qubolite.solving import brute_force

# ===============================
# 0. Config
# ===============================
INPUT_DIR = "Data"
OUTPUT_DIR = "qubolite_result"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ===============================
# 1. Process All Files
# ===============================
for filename in os.listdir(INPUT_DIR):
    input_path = os.path.join(INPUT_DIR, filename)
    
    if os.path.isfile(input_path):
        print(f"Processing: {filename}...")
        
        try:
            # 2. Load QUBO matrix
            Q_mat = np.loadtxt(input_path)
            n = Q_mat.shape[0]

            # 3. Create QUBO object
            Q = qubo(Q_mat)

            # 4. Get the Optimal Solution (using qubolite's brute_force)
            best_x, best_energy = brute_force(Q)

            # 5. Generate ALL possible solutions and calculate energies
            # Note: Q(x) is the correct way to get energy in qubolite
            all_solutions = []
            for bits in itertools.product([0, 1], repeat=n):
                x = np.array(bits)
                energy = Q(x) # Calling the object calculates the energy
                all_solutions.append((x, energy))

            # Sort solutions by energy (lowest to highest)
            all_solutions.sort(key=lambda item: item[1])

            # 6. Save results
            matrix_name = os.path.splitext(filename)[0]
            output_path = os.path.join(OUTPUT_DIR, f"{matrix_name}_results.txt")

            with open(output_path, "w") as f:
                f.write("QUBOLite Search Results\n")
                f.write("========================\n")
                f.write(f"Input file       : {filename}\n")
                f.write(f"Variables (n)    : {n}\n\n")
                
                f.write(f"--- BRUTE FORCE OPTIMUM ---\n")
                f.write(f"Best bitstring   : {best_x.astype(int).tolist()}\n")
                f.write(f"Minimum energy   : {best_energy}\n\n")
                
                f.write(f"--- ALL {2**n} POSSIBLE STATES (Sorted) ---\n")
                f.write(f"{'Bitstring':<{n+2}} | {'Energy'}\n")
                f.write("-" * (n + 15) + "\n")
                
                for x, energy in all_solutions:
                    bit_str = "".join(map(str, x.astype(int)))
                    f.write(f"{bit_str:<{n+2}} | {energy}\n")

            print(f"Done: {filename} (Optimal: {best_energy})")
            
        except Exception as e:
            print(f"Error processing {filename}: {e}")

print("\nBatch processing complete.")
```
