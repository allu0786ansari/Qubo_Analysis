```python
import numpy as np

Q = np.loadtxt('Data/QUBO_1y6r_matrix.txt')
n = Q.shape[0]

with open('Data/matrix_1y6r.qubo', 'w') as f:
    f.write(f"{n}\n")  # n=8 vars at top (matches test.txt style)
    # Triplets 0-indexed i <= j (sprs-safe)
    for i in range(n):
        for j in range(i, n):
            f.write(f"{i} {j} {Q[i,j]/10000:.10f}\n")

print("converted matrix is saved")
```

```python
import hercules

# Read QUBO
problem = hercules.read_qubo("Data/matrix_1y6r.qubo")

# Branch and Bound solve
x_soln, obj, runtime, nodes, lp_solves = hercules.solve_branch_bound(
    problem,
    timeout=10.0,
    sub_problem_solver="clarabel_lp"
)

# Results
print("Exact energy:", obj)
print("Runtime (s):", runtime)
print("B&B nodes explored:", nodes)
print("LP relaxations solved:", lp_solves)

# Binary string
binary_str = ''.join(map(str, x_soln))
print("Binary string:", binary_str)

# Active: 1-based indices
active_indices_1based = [i + 1 for i, bit in enumerate(x_soln) if bit == 1]
print("Active bit indices (1-based):", active_indices_1based)
```

# Approach via the mixed local search heuristic withAnother Approach 

```python
import hercules
import random
import time

# Read QUBO
problem = hercules.read_qubo('Data/matrix_1y6r.qubo')
num_x = problem[-1]  # number of variables

print(f"Problem size: {num_x}")

# Random initial point
random.seed(time.time())
x_0 = [random.randint(0, 1) for _ in range(num_x)]

# Timing
start = time.time()

# Mixed Local Search
x_soln, obj = hercules.mls(problem, x_0, 100)

end = time.time()

# Results
print("Solution vector:", x_soln)
print("Energy:", obj, "in", end - start, "seconds")

# Binary string
binary_str = ''.join(map(str, x_soln))
print("Binary string:", binary_str)

# ✅ Active bit indices (where bit == 1)
active_indices = [i for i, bit in enumerate(x_soln) if bit == 1]
print("Active bit indices:", active_indices)
```
